# Adding a `favicon.ico` to a Web Project

This notebook demonstrates the steps needed to add a favicon icon file to a web application and ensure it is served correctly via HTTP. The sections follow the outline provided by the user.

## 1. Section: Create or Obtain favicon.ico

A favicon is typically a small 16x16 or 32x32 icon. You can create one using an image editor (e.g. GIMP, Photoshop) or use an online generator. Once created, save it as `favicon.ico` in the root of your project directory. For demonstration we’ll simulate this step by creating a placeholder file.

````python
# simulate creating a placeholder favicon file
with open('/workspaces/drakweb/favicon.ico', 'wb') as f:
    f.write(b"\0\0\1\0")  # not a valid icon but placeholder

print("favicon.ico created in project root")
````

## 2. Section: Add Favicon to HTML Head

Edit the HTML template of your application to include a link to the favicon within the `<head>` element. This makes browsers request `/favicon.ico` automatically.

Example snippet (e.g., in `public/index.html` for a React app):

```html
<head>
  <!-- other head tags -->
  <link rel="shortcut icon" href="/favicon.ico" type="image/x-icon" />
</head>
```

We won't modify a real template here but the snippet above shows what to add.

## 3. Section: Serve the Favicon via HTTP Server

The web server must be configured to serve the `favicon.ico` file when a browser requests it. With Express, serving static files from the root will handle this.

Example Express code:

```js
const express = require('express');
const app = express();
app.use(express.static('public')); // assumes favicon.ico in public/
app.listen(3000);
```

For a simple Python HTTP server, you could run:

```bash
cd /path/to/project && python3 -m http.server
```

and ensure `favicon.ico` is in the same directory.

## 4. Section: Verify Favicon Deployment

To check that the file is being served correctly, you can use `curl` or open the site in a browser:

```bash
curl -I http://localhost:3000/favicon.ico
```

Look for a `200 OK` response and `Content-Type: image/x-icon`. Browsers will also display the icon in the tab once cached.

```python
import requests
resp = requests.get('http://localhost:3000/favicon.ico')
print(resp.status_code, resp.headers.get('content-type'))
```

That completes the demonstration of adding and verifying a favicon.
